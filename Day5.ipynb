```python
def sum_all(start, end):
  output=0
  for i in range(start,end+1):
    output+=i
  return output

print('0에서 100까지의 합:',sum_all(0,100))
print('0에서 100까지의 합:',sum_all(0,1000))
print('0에서 100까지의 합:',sum_all(50,100))
print('0에서 100까지의 합:',sum_all(500,1000))
```

```python
#반복문으로 팩토리얼 구하기

def facto(n):
  output = 1
  for i in range(1, n+1):
    output *= i
  return output
  
print('3!:',facto(3))
print('5!:',facto(5))
print('7!:',facto(7))
```

```python
#재귀(recursion)문으로 팩토리얼 구하기

def facto(n):
  if n <= 1:
    return 1
  else:
    return n*facto(n-1)

print('3!:',facto(3))
print('5!:',facto(5))
print('7!:',facto(7))

# 개발자들 사이에서는 재귀함수를 절대 사용하지 말자 라는 의견도 있다.
# 하지만 가독성이 좋고 코드가 간결하고 멋스러워 보이는 장점도 있다. (예쁜 쓰레기)
```

```python
#피보나치 수열 (토끼의 번식 속도, 바이러스 감염 속도...)
def fibo(n):
  if n == 1 or n == 2:
    	return 1
  else:
    	return fibo(n-1) + fibo(n-2)

print('fibonacci(1):',fibo(1))
print('fibonacci(2):',fibo(2))
print('fibonacci(3):',fibo(3))
print('fibonacci(4):',fibo(4))
print('fibonacci(5):',fibo(5))
```

```python
def add_many(*args):
  result = 0

  for arg in args:
    result+=arg
  return result

print(add_many(1,2,3))
print(add_many(1,2,3,4,5))
```

```python
def add_mul(choice,*args):
  if choice =='add':
    result = 0
    for i in args:
      result += i
      
  elif choice == 'mul':
    result = 1
    for i in args:
      result *= i
  return result

print(add_mul('add',1,2,3,4,5))
print(add_mul('mul',1,2,3))
```
