```python
# 1 알고리즘이란....
# - 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어 또는 생각
# - 알고리즘 자체는 일상생활에서도 많이 이미 사용하고 있다.
# - 요리의 레시피, 음악의 악보, 가전제품 사용설명서 등
# - 문자, 기호, 일러스트 등으로 나타낸다.

# 2. 알고리즘과 프로그램
# - 알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다.
# - 프로그래밍 언어(파이썬, 자바, 씨...)는 컴퓨터에 지시하기 위한 인공 언어이다.

# 3. 프로그램 작성에서의 알고리즘
# - 프로그램 작성은 기획, 설계(*), 프로그래밍, 디버그 로 진행한다.
# - 알고리즘은 설계 단계에서 필요하다.

# 4. 좋은 알고리즘이란 ??
# - 알기 쉽다. (쉬운 이해)
# - 속도가 빠르다.
# - 효율적이다.(메모리를 작게 사용한다.)
# - 재이용(범용성)이 쉽다.(객체지향프로그래밍)

# 5. 왜 알고리즘을 공부해야 하나
# - 좋은 프로그램을 만들기위해 (빠르고, 효율적, 범용성)
# - 프로그램의 좋고 나쁨을 판단하기 위해
# - 프로그램 작성 과정 전체를 효율화하기 위해
# - 프로그램 기술을 향상시키기 위해 

# 6. 알고리즘의 조건
# - 정확한 결과를 얻을 수 있어야 한다.
#   (알고리즘은 문제나 과제를 해결하기 위한 절차)
# - 반드시 종료되어야 한다.
#   (중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다.)

# 7. 알고리즘의 기본형 3가지
# - 순차 구조 : 처음부터 순서대로 처리하는 절차
#   예) 심부름
#   1. 슈퍼마켓에 간다.
#   2. 라면을 산다.
#   3. 집으로 돌아온다.
#   실행을 원하는 처리를 위로부터 순서대로 작성한다.
#   알고리즘 중에서 가장 많이 사용된다.

# - 선택 구조 : 조건식으로 판단해 실행할 절차를 전환하는 구조

# - 반복 구조 : 조건을 만족하는 동안 같은 처리를 반복하는 절차
```

```python
#Linear Search 선형 탐색법

#보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다.
#탐색 처리는 반복 구조를 사용하여 기술한다.
#반복 구조에서는 종료 조건을 잊지 말자

#첨자 (방번호) 0 1 2 3 4
#데이터        4 2 3 5 1  찾을 데이터 : 5
```

```python
#풀이1

arr = [4,2,3,5,1]

for i in arr:
  if i == 5:
    print(f'{arr.index(i)}번째 요소가 일치')
    break
  elif i == arr[-1]:
    print(f'찾지 못했습니다')
```

```python
#풀이2

arr = [4,2,3,5,1]

for i in range(0,5):
  if arr[i] == 5:
    print(f'{i}번째 요소가 일치')
    break
  elif i == 5:
    print(f'찾지 못했습니다')
```

```python
def linear_search(no,list):
  for i in range(len(list)):
    if no == list[i]:
      return print(f'{i}번째 요소가 일치')
    return print(f'찾지 못했습니다')

print(linear_search(5, [4,2,3,5,1]))
```

```python
def test():
  print("test()함수의 첫 줄입니다.")
  try:
    print('try 구문이 실행되었습니다.')
    return
    print('try 구문의 return 아래입니다.')
  except:
    print('except구문이 실행되었습니다.')
  else:
    print('else구문이 실행되었습니다.')
  finally:
    print('finally구문이 실행되었습니다.')
  print('test()함수의 마지막 줄입니다.')


test()
```
